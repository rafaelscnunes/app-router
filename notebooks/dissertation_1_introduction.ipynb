```python
import os
import pandas as pd
import matplotlib.pyplot as plt

WORKING_FOLDER = 'var/cache/'

pedidos = {
    'Ano': [1999, 2002, 2005, 2008, 2011, 2014, 2015, 2016, 2017, 2018],
    'Todos': [19584, 20231, 21729, 26512, 31792, 33075, 32936, 30946, 28576, 27444],
    'Residentes': [6137, 6954, 7229, 7595, 7714, 7296, 7247, 8014, 8323, 7473]
}
# fonte: Relat√≥rio de Atividades 2018
```

```python
df = pd.DataFrame(pedidos)
df
```

```python
fig, axs = plt.subplots(figsize=(20, 10))
df.plot.bar(x='Ano', y=['Todos', 'Residentes'], figsize=(20,10), rot=45, colormap='tab20', sort_columns=True, title='Pedidos de patentes depositados no INPI', ax=axs);
for p in axs.patches:
    axs.annotate(str(p.get_height()), (p.get_x()+p.get_width()/2, p.get_height()*1.005), ha='center')
fig.savefig(os.path.join(WORKING_FOLDER, f'pedidos_patentes_2018.png'), bbox_inches='tight');
```
